```python
n = int(input("Quantos registros serão lidos? "))

soma_m = soma_f = 0
qtd_m = qtd_f = 0

for i in range(n):
    sexo = input("Digite o sexo (M/F): ").upper()
    idade = int(input("Digite a idade: "))
    if sexo == "M":
        soma_m += idade
        qtd_m += 1
    elif sexo == "F":
        soma_f += idade
        qtd_f += 1

if qtd_m > 0:
    print(f"Média de idades dos homens = {soma_m / qtd_m:.2f}")
else:
    print("Nenhum homem informado.")

if qtd_f > 0:
    print(f"Média de idades das mulheres = {soma_f / qtd_f:.2f}")
else:
    print("Nenhuma mulher informada.")
```

```python
maior = float('-inf')
menor = float('inf')

for i in range(10):
    num = float(input(f"Digite o {i+1}º número real: "))
    if num > maior:
        maior = num
    if num < menor:
        menor = num

print(f"O maior número é {maior}")
print(f"O menor número é {menor}")
```

```python
valores = []

while True:
    num = float(input("Digite um número real (0 para parar): "))
    if num == 0:
        break
    valores.append(num)

valores.sort()
print("Números em ordem crescente:")
for v in valores:
    print(v)
```
